```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Chi2 veri setini yükleyin
chi2_train_df = pd.read_csv('updated_data/chi2_training.csv')
chi2_test_df = pd.read_csv('updated_data/chi2_testing.csv')

# Eğitim verileri için öznitelikler ve hedef değişkeni ayırın
X_train = chi2_train_df.drop('target', axis=1)
y_train = chi2_train_df['target']

# Test verileri için öznitelikler ve hedef değişkeni ayırın
X_test = chi2_test_df.drop('target', axis=1)
y_test = chi2_test_df['target']

correlation_train_df = pd.read_csv('updated_data/correlation_training.csv')
correlation_test_df = pd.read_csv('updated_data/correlation_testing.csv')

# Eğitim verileri için öznitelikler ve hedef değişkeni ayırın
X_train1 = correlation_train_df.drop('target', axis=1)
y_train1 = correlation_train_df['target']

# Test verileri için öznitelikler ve hedef değişkeni ayırın
X_test1 = correlation_test_df.drop('target', axis=1)
y_test1 = correlation_test_df['target']


# Lasso veri setini yükleyin
lasso_train_df = pd.read_csv('updated_data/lasso_training.csv')
lasso_test_df = pd.read_csv('updated_data/lasso_testing.csv')

# Eğitim verileri için öznitelikler ve hedef değişkeni ayırın
X_train2 = lasso_train_df.drop('target', axis=1)
y_train2 = lasso_train_df['target']

# Test verileri için öznitelikler ve hedef değişkeni ayırın
X_test2 = lasso_test_df.drop('target', axis=1)
y_test2 = lasso_test_df['target']
```

```python
# Decision Tree modelini oluşturun ve eğitin
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)

# Eğitim verileri üzerinde tahmin yapın
y_train_pred = tree_model.predict(X_train)

# Test verileri üzerinde tahmin yapın
y_test_pred = tree_model.predict(X_test)

# Performans metriklerini hesaplayın
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
train_precision = precision_score(y_train, y_train_pred)
test_precision = precision_score(y_test, y_test_pred)
train_recall = recall_score(y_train, y_train_pred)
test_recall = recall_score(y_test, y_test_pred)
train_f1 = f1_score(y_train, y_train_pred)
test_f1 = f1_score(y_test, y_test_pred)

print(f'Decision Trees - Eğitim Verileri - Accuracy: {train_accuracy}')
print(f'Decision Trees - Eğitim Verileri - Precision: {train_precision}')
print(f'Decision Trees - Eğitim Verileri - Recall: {train_recall}')
print(f'Decision Trees - Eğitim Verileri - F1 Score: {train_f1}')

print(f'Decision Trees - Test Verileri - Accuracy: {test_accuracy}')
print(f'Decision Trees - Test Verileri - Precision: {test_precision}')
print(f'Decision Trees - Test Verileri - Recall: {test_recall}')
print(f'Decision Trees - Test Verileri - F1 Score: {test_f1}')

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Tahmin Edilen')
plt.ylabel('Gerçek')
plt.title('Confusion Matrix - Decision Trees (Chi2)')
plt.show()
```

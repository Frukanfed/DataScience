```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
chi2_train_df = pd.read_csv('feature_selected_data/anova_train.csv')
chi2_test_df = pd.read_csv('feature_selected_data/anova_test.csv')

X_train = chi2_train_df.drop('Recurred', axis=1)
y_train = chi2_train_df['Recurred']

X_test = chi2_test_df.drop('Recurred', axis=1)
y_test = chi2_test_df['Recurred']

correlation_train_df = pd.read_csv('feature_selected_data/correlation_train.csv')
correlation_test_df = pd.read_csv('feature_selected_data/correlation_test.csv')

X_train1 = correlation_train_df.drop('Recurred', axis=1)
y_train1 = correlation_train_df['Recurred']

X_test1 = correlation_test_df.drop('Recurred', axis=1)
y_test1 = correlation_test_df['Recurred']

lasso_train_df = pd.read_csv('feature_selected_data/rfe_train.csv')
lasso_test_df = pd.read_csv('feature_selected_data/rfe_test.csv')

X_train2 = lasso_train_df.drop('Recurred', axis=1)
y_train2 = lasso_train_df['Recurred']

X_test2 = lasso_test_df.drop('Recurred', axis=1)
y_test2 = lasso_test_df['Recurred']
```

```python
# KNN modelini oluşturun ve eğitin
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

y_train_pred = knn_model.predict(X_train)
y_test_pred = knn_model.predict(X_test)

# Performans metriklerini hesaplayın
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
train_precision = precision_score(y_train, y_train_pred)
test_precision = precision_score(y_test, y_test_pred)
train_recall = recall_score(y_train, y_train_pred)
test_recall = recall_score(y_test, y_test_pred)
train_f1 = f1_score(y_train, y_train_pred)
test_f1 = f1_score(y_test, y_test_pred)

print(f'KNN - Training - Accuracy: {train_accuracy}')
print(f'KNN - Training - Precision: {train_precision}')
print(f'KNN - Training - Recall: {train_recall}')
print(f'KNN - Training - F1 Score: {train_f1}')

print(f'KNN - Testing - Accuracy: {test_accuracy}')
print(f'KNN - Testing - Precision: {test_precision}')
print(f'KNN - Testing - Recall: {test_recall}')
print(f'KNN - Testing - F1 Score: {test_f1}')

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Tahmin Edilen')
plt.ylabel('Gerçek')
plt.title('Confusion Matrix  KNN ChiSquare')
plt.show()
```
